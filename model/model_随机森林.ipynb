```python
#导入模块
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split  
from sklearn.ensemble import RandomForestRegressor  
from sklearn.metrics import mean_squared_error, r2_score    
from sklearn.preprocessing import StandardScaler  

#导入文件
train_housing = pd.read_csv("train.csv", encoding = "UTF-8")
test_housing = pd.read_csv("test.csv", encoding = "UTF-8")

#添加截距
train_housing['cons'] = 1
test_housing['cons'] = 1
print(train_housing.head())

#显示行项
print(train_housing.columns)

#清洗缺失值
train_housing = train_housing.dropna()
test_housing = train_housing.dropna()
train_housing

# 划分数据集为训练集和测试集  
X_train = train_housing[['cons', 'houseSpace', 'fitment', 'houseFloor', 'houseStructure', 'isNearSubway', 'buildTime',  'longitude', 'latitude']]  
y_train = train_housing["unitPrice"]  
X_test = test_housing[['cons', 'houseSpace', 'fitment', 'houseFloor', 'houseStructure', 'isNearSubway', 'buildTime',  'longitude', 'latitude']] 

# 特征缩放
scaler = StandardScaler()  
X_train_scaled = scaler.fit_transform(X_train)  
X_test_scaled = scaler.transform(X_test)  

# 训练随机森林模型 
rf = RandomForestRegressor(n_estimators=100, random_state=42)  
rf.fit(X_train_scaled, y_train) 

# 预测测试集  
y_pred = rf.predict(X_test_scaled)  

# 评估模型  
mse = mean_squared_error(test_housing["unitPrice"], y_pred)  
rmse = mse ** 0.5  
r2 = r2_score(test_housing["unitPrice"], y_pred) 

print(f"Mean Squared Error: {mse}")  
print(f"Root Mean Squared Error: {rmse}")  
print(f"R2 Score: {r2}")

# 查看特征重要性
feature_names = ['cons', 'houseSpace', 'fitment', 'houseFloor', 'houseStructure', 'isNearSubway', 'buildTime', 'longitude', 'latitude']
importances = rf.feature_importances_ 
feature_importances = dict(zip(feature_names, importances)) 
sorted_importances = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True) 

feature_importance_df = pd.DataFrame(sorted_importances, columns=['Feature', 'Importance'])  
print(feature_importance_df)

# 交叉验证检测过拟合问题
from sklearn.model_selection import cross_val_score  

# 使用Pipeline来确保缩放是在每个折叠内部进行的 
from sklearn.pipeline import make_pipeline  

pipeline = make_pipeline(scaler, rf)  

# 执行5折交叉验证 
cv_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='neg_mean_squared_error') 

# 将MSE的负值转回正值  
mse_scores = -cv_scores 

# 计算平均MSE和RMSE  
mean_mse = mse_scores.mean()  
rmse = mean_mse ** 0.5  

# 交叉验证计算R2
from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(pipeline, X_train, y_train cv=5)  
r2_scores = cross_val_score(pipeline, X_train, y_train, cv=5, scoring='r2')  
mean_r2 = r2_scores.mean()  

# 打印结果  
print(f"mean_mse: {mean_mse}")  
print(f"rmse: {rmse}") 
print(f"mean_r2: {mean_r2}")
```
